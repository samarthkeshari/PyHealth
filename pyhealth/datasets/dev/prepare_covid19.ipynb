This script prepares the COVID-19 dataset for use in the pyhealth package.

URL: https://www.kaggle.com/datasets/tawsifurrahman/covid19-radiography-database

```python
import os
import sys
import pandas as pd
import json
```

```python
raw_data_path = "/srv/local/data/zw12/raw_data/covid19-radiography-database/COVID-19_Radiography_Dataset"
```

```python
!ls $raw_data_path
```

```python
covid = pd.DataFrame(pd.read_excel(f"{raw_data_path}/COVID.metadata.xlsx"))
covid
```

```python
print(covid.FORMAT.unique())
print(covid.SIZE.unique())
```

```python
covid["FILE NAME"] = covid["FILE NAME"].apply(lambda x: f"COVID/images/{x}.png")
covid["label"] = "COVID"
covid
```

```python
lung_opacity = pd.DataFrame(pd.read_excel(f"{raw_data_path}/Lung_Opacity.metadata.xlsx"))
lung_opacity
```

```python
print(lung_opacity.FORMAT.unique())
print(lung_opacity.SIZE.unique())
```

```python
lung_opacity["FILE NAME"] = lung_opacity["FILE NAME"].apply(lambda x: f"Lung_Opacity/images/{x}.png")
lung_opacity["label"] = "Lung Opacity"
lung_opacity
```

```python
normal = pd.DataFrame(pd.read_excel(f"{raw_data_path}/Normal.metadata.xlsx"))
normal
```

```python
print(normal.FORMAT.unique())
print(normal.SIZE.unique())
```

```python
normal["FILE NAME"] = normal["FILE NAME"].apply(lambda x: x.capitalize())
normal
```

```python
normal["FILE NAME"] = normal["FILE NAME"].apply(lambda x: f"Normal/images/{x}.png")
normal["label"] = "Normal"
normal
```

```python
viral_pneumonia = pd.DataFrame(pd.read_excel(f"{raw_data_path}/Viral Pneumonia.metadata.xlsx"))
viral_pneumonia
```

```python
print(viral_pneumonia.FORMAT.unique())
print(viral_pneumonia.SIZE.unique())
```

```python
viral_pneumonia["FILE NAME"] = viral_pneumonia["FILE NAME"].apply(lambda x: f"Viral Pneumonia/images/{x}.png")
viral_pneumonia["label"] = "Viral Pneumonia"
viral_pneumonia
```

```python
df = pd.concat([covid, lung_opacity, normal, viral_pneumonia], axis=0, ignore_index=True)
df
```

```python
df = df.drop(columns=["FORMAT", "SIZE"])
df.columns = ["path", "label", "url"]
df
```

```python
for path in df.path:
    assert os.path.isfile(os.path.join(raw_data_path, path))
```

```python
df.to_json(f"{raw_data_path}/metadata.jsonl", orient="records", lines=True)
```
